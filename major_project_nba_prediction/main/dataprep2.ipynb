### dataprep2.ipynb
Use output file from dataprep.ipynb to create dataset for making predictions

```python
import pandas as pd
import numpy as np

# read file
df = pd.read_csv('../output_files/dataprep_output.csv').drop(columns=["DATE", "HOME_TEAM_ABR", "AWAY_TEAM_ABR"])

df_random = df.copy()
df_fixed = df.copy()

df
```

Remove outliers based on z-score for table values

```python
# file 1: replace NaN values in 'PTS_TOTAL_SCOREBOOK' with random between min and max

import random

na_mask = df_random['PTS_TOTAL_SCOREBOOK'].isna()
min_num = 220
max_num = 240
to_replace = [random.randint(min_num, max_num) for _ in range(na_mask.sum())]
df_random.loc[na_mask, 'PTS_TOTAL_SCOREBOOK'] = to_replace

# set classifier y value (whether actual total is over or under sb total)
df_random = df_random.assign(PTS_TOTAL_OVER_SB = df_random['PTS_TOTAL'] > df_random['PTS_TOTAL_SCOREBOOK']);

df_random = df_random.dropna()

df_random
```

```python
# file 2: replace all values with a fixed number

# set fixed number as median points total for each game
fixed_num = df['PTS_TOTAL'].median(skipna=True)#

df_fixed['PTS_TOTAL_SCOREBOOK'] = fixed_num

# set classifier y value (whether actual total is over or under sb total)
df_fixed = df_fixed.assign(PTS_TOTAL_OVER_SB = df_fixed['PTS_TOTAL'] > df_fixed['PTS_TOTAL_SCOREBOOK']);

df_fixed = df_fixed.dropna()

# percentage of y value
print(df_fixed['PTS_TOTAL_OVER_SB'].value_counts( normalize=True).mul(100).astype(str)+'%')

df_fixed
```

```python
# file 3: keep dataframe as normal and remove all NaN values

# set classifier y value (whether actual total is over or under sb total)
df = df.assign(PTS_TOTAL_OVER_SB = df['PTS_TOTAL'] > df['PTS_TOTAL_SCOREBOOK']);

df = df.dropna()

df
```

```python
#export to .csv file
df_random.to_csv('../output_files/prediction_data_replace_random_sb.csv', encoding='utf-8', index=False)
df_fixed.to_csv('../output_files/prediction_data_fixed_sb.csv', encoding='utf-8', index=False)
df.to_csv('../output_files/prediction_data_unreplaced.csv', encoding='utf-8', index=False)
```
