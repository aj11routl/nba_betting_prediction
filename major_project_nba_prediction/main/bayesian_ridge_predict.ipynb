```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

'''
split into testing and training data

PTS_TOTAL is the outcome column
'''

csv_input = input('Which set should be used in predicting?\n 1.Normal data \n 2.Replace missing scorebook values with random \n 3.Replace all scorebook values with fixed number?');

if csv_input == '1':
    csv_string = 'prediction_data_unreplaced.csv';
elif csv_input == '2':
    csv_string = 'prediction_data_random_sb.csv';
elif csv_input == '3':
    csv_string = 'prediction_data_fixed_sb.csv';
else:
    csv_string = '';
    
df = pd.read_csv('../output_files/' + csv_string).dropna(axis=1).dropna(axis=0)

remove_cols = ['PTS_TOTAL_OVER_SB', 'PTS_TOTAL'];

X = df.drop(columns=remove_cols)

y = df['PTS_TOTAL']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

df
```

```python
X
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error, r2_score
```

## Bayesian Ridge

```python
from sklearn import linear_model

BayReg = linear_model.BayesianRidge()
BayReg.fit(X_train, y_train)

pred_br = BayReg.predict(X_test)

print("MAE (mean absolute error):\t", mean_absolute_error(y_test, pred_br));
print("MSE (mean squared error):\t", mean_squared_error(y_test, pred_br));
print("MAPE (mean absolute % error):\t", mean_absolute_percentage_error(y_test, pred_br)*100 , "%");
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold 

k_folds = KFold(n_splits = 5)

scores = cross_validate(BayReg, X, y, cv=k_folds,
                        scoring=('r2', 'neg_mean_absolute_error',  'neg_mean_absolute_percentage_error', 'neg_mean_squared_error')
                       )

r2_scores = scores['test_r2']
neg_mae_scores = scores['test_neg_mean_absolute_error']
neg_mape_scores = scores['test_neg_mean_absolute_percentage_error']
neg_mse_scores = scores['test_neg_mean_squared_error']

print('\nTest R2 scores:')
print(r2_scores)
print('Average R2 score:')
print(r2_scores.mean())

print('\nTest negative MAE scores:')
print(neg_mae_scores)
print('Average negative MAE score:')
print(neg_mae_scores.mean())

print('\nTest negative MAPE scores:')
print(neg_mape_scores)
print('Average negative MAPE score:')
print(neg_mape_scores.mean())

print('\nTest negative MSE scores:')
print(neg_mse_scores)
print('Average negative MSE score:')
print(neg_mse_scores.mean())

```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, f_regression

all_features = X.columns;

# select top k features
k = 8
selector = SelectKBest(score_func=f_regression, k=k)
X_selected = selector.fit_transform(X, y)

selected_indices = np.argsort(selector.scores_)[::-1][:k]
selected_features = X.columns[selected_indices]

# print selected features
print("\nSelected features using f_regression:")
for val in selected_features:
    print(val)
    
print("\nRemoved features:")
for val in list(set(all_features) - set(selected_features)):
    print(val)

# create new X dataframe with only selected features
X_fs = X[selected_features]
```
